### IA Dentista Lista de Diccionarios

```python
# =====================================================================
# SIMULADOR: IA GESTIÓN Y DIAGNÓSTICO DENTAL
# =====================================================================

# Al inicio, la lista de diccionarios empieza completamente vacía
los_dentistas = []

print("====================================================")
print("🤖 ¡Bienvenido al Sistema Inteligente 'Clínica Dental Pro'! 🦷")
print("====================================================")

while True:
    # MENÚ PRINCIPAL
    print("\n--- MENÚ DE CONTROL ---")
    print("1. Registrar nuevo Dentista y su Especialidad")
    print("2. Ver catálogo de Dentistas (Lista de Diccionarios)")
    print("3. Consultar al Asistente de IA (Diagnóstico Rápido)")
    print("4. Salir del sistema")

    opcion = input("\nSelecciona una opción (1-4): ")

    # OPCIÓN 1: Registrar con Diccionario Vacío al inicio
    if opcion == "1":
        print("\n--- REGISTRO DE NUEVO DENTISTA ---")

        # El diccionario inicia vacío en cada registro
        nuevo_dentista = {}

        # Solicitamos los datos al usuario
        nombre = input("Nombre del odontólogo(a): ")
        especialidad = input("Especialidad (Ej: General, Endodoncia, Ortodoncia): ")
        turno = input("Turno de atención (Matutino / Vespertino): ")

        # Llenamos el diccionario dinámicamente con las llaves y valores
        nuevo_dentista["nombre"] = nombre
        nuevo_dentista["especialidad"] = especialidad
        nuevo_dentista["turno"] = turno

        # Agregamos el diccionario a la lista 'los_dentistas'
        los_dentistas.append(nuevo_dentista)
        print("¡Dentista registrado y guardado con éxito en la base de datos!")

    # OPCIÓN 2: Mostrar los datos usando Ciclos FOR
    elif opcion == "2":
        print("\n--- NUESTRO PERSONAL MÉDICO REGISTRADO ---")
        if len(los_dentistas) == 0:
            print("La lista está vacía. No hay dentistas registrados todavía.")
        else:
            # Primer ciclo FOR: Recorre cada elemento (que es un diccionario) dentro de la lista
            for dentista in los_dentistas:
                print("\n------------------------------------")
                # Segundo ciclo FOR: Recorre las llaves y valores del diccionario actual
                for clave, valor in dentista.items():
                    # Formateamos el texto para que se vea estético (Clave con inicial mayúscula)
                    print(clave.capitalize() + ": " + valor)
            print("------------------------------------")

    # OPCIÓN 3: Módulo de Inteligencia Artificial (Árbol de decisiones con IF-ELIF-ELSE)
    elif opcion == "3":
        print("\n--- ASISTENTE INTELIGENTE DE DIAGNÓSTICO ---")
        print("Describe el problema del paciente seleccionando una opción:")
        print("1. Tiene los dientes chuecos o mala mordida.")
        print("2. Le duele mucho el interior de un diente al tomar cosas frías.")
        print("3. Solo requiere una revisión general o limpieza de rutina.")

        problema = input("Selecciona (1, 2 o 3): ")

        # La "IA" analiza el síntoma y busca en la lista de diccionarios si hay alguien que pueda atenderlo
        if problema == "1":
            print("\n🤖 [IA Analizando]: El paciente requiere tratamiento de ORTODONCIA (Brackets).")
            print("🔍 Buscando especialistas disponibles en el sistema...")

            encontrado = False
            for dentista in los_dentistas:
                if dentista["especialidad"].lower() == "ortodoncia":
                    print("👉 Sugerencia de IA: El/La Dr(a). " + dentista["nombre"] + " está disponible en el turno " + dentista["turno"] + ".")
                    encontrado = True

            if not encontrado:
                print("❌ Alerta: No tenemos ningún Ortodontista registrado en el sistema actualmente.")

        elif problema == "2":
            print("\n🤖 [IA Analizando]: El síntoma apunta a una inflamación del nervio. Requiere ENDODONCIA.")
            print("🔍 Buscando especialistas disponibles en el sistema...")

            encontrado = False
            for dentista in los_dentistas:
                if dentista["especialidad"].lower() == "endodoncia":
                    print("👉 Sugerencia de IA: El/La Dr(a). " + dentista["nombre"] + " está disponible en el turno " + dentista["turno"] + ".")
                    encontrado = True

            if not encontrado:
                print("❌ Alerta: No tenemos ningún Endodoncista registrado en el sistema actualmente.")

        elif problema == "3":
            print("\n🤖 [IA Analizando]: Caso de odontología GENERAL o preventiva.")
            print("🔍 Buscando especialistas disponibles en el sistema...")

            encontrado = False
            for dentista in los_dentistas:
                if dentista["especialidad"].lower() == "general":
                    print("👉 Sugerencia de IA: El/La Dr(a). " + dentista["nombre"] + " puede atenderlo en el turno " + dentista["turno"] + ".")
                    encontrado = True

            if not encontrado:
                print("❌ Alerta: No hay dentistas generales en el sistema. Intenta buscando otro perfil.")

        else:
            print("\n❌ Opción de síntoma no válida.")

    # OPCIÓN 4: Salir usando break debido al While True
    elif opcion == "4":
        print("\nCerrando el sistema dental inteligente... ¡Hasta luego! 🦷🤖")
        break

    else:
        print("\n❌ Opción inválida. Elige un número del 1 al 4.")

print("\n====================================================")
print("Programa finalizado. (Estructura de Datos: Listas + Diccionarios)")
print("====================================================")
```
